```python
import pandas as pd
import prince
dataset = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/balloons/adult+stretch.data')
dataset.columns = ['Color_das', 'Size_das', 'Action_das', 'Age_das', 'Inflated_das']
dataset.head()
```

```python
one_hot = pd.get_dummies(dataset)
display(one_hot)
one_hot.to_csv('test_for_MCA.csv',index=True) 
```

```python
mca_no_one_hot = prince.MCA(one_hot=False)
mca_no_one_hot = mca_no_one_hot.fit(one_hot)


mca_no_one_hot.plot(
    one_hot,
    x_component=0,
    y_component=1,
    show_column_markers=True,
    show_row_markers=True,
    show_column_labels=False,
    show_row_labels=False
)
```
